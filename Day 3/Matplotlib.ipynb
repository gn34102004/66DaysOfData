```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# plotting using matplotlib

# plt scatter

plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
# plt plot

plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'b*',linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
# Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'g--')
plt.subplot(2,2,2)
plt.plot(x,y,'r*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y--')
```

```python
# Compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4 *np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")

# Plot the points using matplotlib

plt.plot(x,y)
plt.show()
```

```python
# Subplot()
# Compute the x and y coordinates for points on sine and cosine curves

x = np.arange(0, 5 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a dubplot grid that has height 2 and width 1,
# and set the first such subplot as active
plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title('Cosine')

# Show the figure 
plt.show()
```

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

## Box Plot using Matplotlib

helps to find percentile, taking normal distribution between the lower value and the standard distribution value and the step size is 100, std will loop from the given range

```python
data = [np.random.normal(0, std, 100) for std in range (1,4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
data
```

```python
data = [np.random.normal(0, std, 100) for std in range (1,4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```

```python
data = [np.random.normal(0, std, 100) for std in range (1,4)]

# rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False);
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)    #explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
```
