## Performance Metrics On MultiClass Classification Problems

```python
from sklearn import metrics
```

```python
C="Cat"
D="Dog"
F="Fox"
```

![Screenshot 2024-02-14 at 12.10.24â€¯AM.png](<attachment:Screenshot 2024-02-14 at 12.10.24â€¯AM.png>)

```python
# True values
y_true = [C,C,C,C,C,C,   F,F,F,F,F,F,F,F,F,F,   D,D,D,D,D,D,D,D,D]
 # Predicted values
y_pred = [C,C,C,C,D,F,   C,C,C,C,C,C,D,D,F,F,   C,C,C,D,D,D,D,D,D]

# Print the confusion matrix
print(metrics.confusion_matrix(y_true, y_pred))

# Print the precision and recall, among the metrices
print(metrics.classification_report(y_true,y_pred,digits=3))
```
