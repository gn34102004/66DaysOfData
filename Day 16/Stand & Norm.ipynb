# Normalization vs Standardization

## Normalization (min-max Normalization)

In this approach we will scale down the values of the features between 0 to 1.

![Screenshot 2024-02-08 at 10.34.32â€¯PM.png](<attachment:Screenshot 2024-02-08 at 10.34.32â€¯PM.png>)

```python
import pandas as pd 
```

```python
df=pd.read_csv("wine_data.csv", header=None,usecols=[0,1,2])
df.columns=['Class','Alcohol','Malic']
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling=MinMaxScaler()  #object
```

```python
scaling.fit_transform(df[['Alcohol','Malic']])
```

## Standardization(Z-Score Normalization)

Here all the features will be transformed in such a way that it will have the properties of a standard normal distribution with mean=0 and standard deviation=1

![Screenshot 2024-02-08 at 10.34.55â€¯PM.png](<attachment:Screenshot 2024-02-08 at 10.34.55â€¯PM.png>)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaling=StandardScaler()
```

```python
scaling.fit_transform(df[['Alcohol','Malic']])
```
